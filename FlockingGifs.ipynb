```python
import numpy as np
import matplotlib.pyplot as plt
import imageio
```

```python
x = np.loadtxt('datos_x.txt')
y = np.loadtxt('datos_y.txt')
vx = np.loadtxt('datos_vx.txt')
vy = np.loadtxt('datos_vy.txt')
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x[0],y[0],color='g')
plt.show()
```

```python
images = []

for i in range(len(x)):
    plt.figure(figsize=(8,8))
    #plt.scatter(x[i][:50],y[i][:50],color='r')
    plt.scatter(x[i],y[i],color='g')
    plt.quiver(x[i],y[i],vx[i],vy[i],headlength=5)
    plt.axis([0., 1., 0., 1.])
    plt.savefig("Fig{0}".format(i))
    plt.close()
    images.append(imageio.imread("Fig{0}.png".format(i)))

imageio.mimsave('flocking.gif', images, 'GIF', duration=0.03)
```
